```python
from __future__ import print_function
from __future__ import division

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import datasets

iris = datasets.load_iris()
print(type(iris))
print(iris.keys())

sns.set(style="white", color_codes=True)
```

```python
data = iris.data
target = iris.target
feature_names = iris.feature_names
print('data.shape=%s' % str(data.shape))
print('target.shape=%s' % str(target.shape))
print(feature_names)

# 导入数据
# sepal: 萼片   petal：花瓣
data_xy = np.hstack([data, target.reshape([-1, 1])])
df_iris = pd.DataFrame(data_xy, columns=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species'])
df_iris.head()
```

统计类别分布

```python
df_iris['Species'].value_counts()
```

## 散点图

```python
# 绘制所有点的分布

df_iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
plt.show()
```

```python
# 在上图基础上加上直方图

sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df_iris, size=5)
plt.show()
```

```python
# 在 plot 的时候加上 kind=‘kde’ 就能够绘制出相应的概率密度分布图
# kde: kernel density estimation 核密度估计
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df_iris, size=5, kind='kde')
plt.show()
```

## 直方图

```python
sns.set_style('white')
sns.distplot(df_iris.SepalLengthCm.values, 
             bins=20,
             kde=False,
             norm_hist=True,  # 是否归一化处理
             rug=True);
plt.show()
```

## HexBin 图
上述直方图用的是单变量，如果是双变量，可以使用 HexBin 图

```python
with sns.axes_style("white"):
    sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df_iris, kind='hex', color='k', size=3)
    plt.show()
```

## 分类数据绘制
FacetGrid可以绘制最多三个维度：row，col和hue。前两者与所得轴数有明显的对应关系;将hue变量视为沿着深度轴的第三维，其中不同的级别用不同的颜色绘制。

```python
# hue 色调，即根据 Species 来进行上色
fig = sns.FacetGrid(df_iris, hue="Species", 
                    hue_kws={"marker": ['^', '+', '*']},
                    size=5)
# 利用 map 函数， 对 plt.scatter 的输出进行上色
fig.map(plt.scatter, "SepalLengthCm", "SepalWidthCm")
# 添加 legend
fig.add_legend()
plt.show()
```

## 核密度估计

```python
# 同理，使用上面的上色方式来绘制密度分布
fig = sns.FacetGrid(df_iris, hue="Species", size=5)
fig.map(sns.kdeplot, "SepalLengthCm",shade=True)
fig.add_legend()
plt.show()
```

```python
fig = sns.jointplot(x="SepalLengthCm", y="SepalWidthCm",data=df_iris, 
                    kind="kde", 
                    color="m")
fig.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
fig.ax_joint.collections[0].set_alpha(0)
fig.set_axis_labels("$SepalLengthCm$", "$SepalWidthCm$")
plt.show()
```

## 箱线图

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=df_iris)
plt.show()
```

```python
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=df_iris)
# sns.stripplot(x, y) 分类散点图， 根据 y 分类绘制散点图
# 设置 jitter=True 使所有的点不落在一条线上
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=df_iris, jitter=True, edgecolor="gray")
plt.show()
```

## 提琴图
提琴图结合了上面箱线图和核密度估计图的优点，把二者直接整合在一起

```python
ax = sns.violinplot(x="Species", y="PetalLengthCm", data=df_iris, size=6)
ax.set_xlabel('label')
plt.show()
```

## pairplot

```python
# 这个函数非常慢，只适合做小数据量的分析

sns.pairplot(df_iris, hue="Species", size=3)
plt.show()
```

```python
sns.pairplot(df_iris, hue="Species", palette="Set2", diag_kind="kde", size=2.5)
plt.show()
```

## 样式控制
[\[参考\]Seaborn(sns)官方文档学习笔记（第一章 艺术化的图表控制）](https://zhuanlan.zhihu.com/p/27435863)

### 主题控制
有5个seaborn的主题：

- darkgrid 黑色网格（默认）
- whitegrid 白色网格
- dark 黑色背景
- white 白色背景
- ticks 应该是四周都有刻度线的白背景

```python
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for i in range(5):
    sns.set_style(styles[i])
    fig = sns.FacetGrid(df_iris, hue="Species", size=2.5)
    fig.map(sns.kdeplot, "SepalLengthCm",shade=True)
    fig.add_legend()
    plt.title(str(i) + '-' + styles[i])
    plt.show()
```

```python
# 查看样式的参数
sns.axes_style()
```

```python
# 直接设置参数值进行调整 

sns.set_style("darkgrid", {"axes.facecolor": ".5"})
fig = sns.FacetGrid(df_iris, hue="Species", size=5)
fig.map(sns.kdeplot, "SepalLengthCm",shade=True)
fig.add_legend()
plt.title(str(i) + '-' + styles[i])
plt.show()
```
