## 基本模型

1.导入相关的包

```python
import numpy as np
import pandas as pd 
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import time
```

2.数据分析

```python
# 导入数据,一共有 54 列数据，其中最后一列表示 label（评论数）
df_dataset = pd.read_csv('data/Dataset/Training/Features_Variant_1.csv', delimiter=',', names=['f_' + str(i) for i in range(53)] + ['comment'])
print('sample number = %d' % len(df_dataset))
df_dataset.head()
```

```python
comment_count = df_dataset.comment.value_counts()
print(comment_count)
```

评论数的分布

```python
import seaborn as sns
```

```python
sns.violinplot(y='comment', data=df_dataset)
plt.show()
```

```python
# 测试数据,直接从训练集中划分部分数据出来作为验证集
seed = 7
test_size = 0.33
X = df_dataset.values[:, :-1]
y = df_dataset.values[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

### 输入数据的形式
处理成 xgb.DMatrix 格式

3.训练模型

[XGBoost-Python完全调参指南-参数解释篇](http://blog.csdn.net/wzmsltw/article/details/50994481)

objective [ default=reg:linear ] 
定义学习任务及相应的学习目标，可选的目标函数如下：
- “reg:linear” –线性回归。
- “reg:logistic” –逻辑回归。
- “binary:logistic” –二分类的逻辑回归问题，输出为概率。
- “binary:logitraw” –二分类的逻辑回归问题，输出的结果为wTx。
- “count:poisson” –计数问题的poisson回归，输出结果为poisson分布。
- 在poisson回归中，max_delta_step的缺省值为0.7。(used to safeguard optimization)
- “multi:softmax” –让XGBoost采用softmax目标函数处理多分类问题，同时需要设置参数num_class（类别个数）
- “multi:softprob” –和softmax一样，但是输出的是ndata * nclass的向量，可以将该向量reshape成ndata行nclass列的矩阵。每行数据表示样本所属于每个类别的概率。
- “rank:pairwise” –set XGBoost to do ranking task by minimizing the pairwise loss

baseline: 统计所有结果的平均值作为最后的输出

```python
from sklearn.metrics import  mean_absolute_error
```

```python
mean_train = np.mean(y_train)
baseline_predictions = np.ones_like(y_test) * mean_train

mae_baseline = mean_absolute_error(y_test, baseline_predictions)
print('Baseline MAE is {:.4f}'.format(mae_baseline))
```

使用 xgb 训练模型进行预测

```python
tic = time.time()
model = XGBRegressor(n_estimators=200, silent=False)  # 全部使用默认参数简单训练一个模型
model.fit(X_train, y_train)
print('time cost %gs' % (time.time() - tic))
```

```python
xgb_predictions = model.predict(X_test)
mae_xgb = mean_absolute_error(y_test, xgb_predictions)
print('Baseline MAE is {:.4f}'.format(mae_xgb))
```

```python
y_leaf = model.apply(X_test)
print('y_leaf.shape=', y_leaf.shape)
# 树的棵数
min_id = np.min(y_leaf, axis=0)
max_id = np.max(y_leaf, axis=0)
min_max_id = zip(min_id, max_id)
print('min_max_id=%s' % str(min_max_id))
```
