## 基本模型

1.导入相关的包

```python
import numpy as np
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import time
```

2.导入数据，分开特征和标签，然后划分训练集和测试集

```python
# 糖尿病数据，最后一列是类别
dataset = np.loadtxt('data/pima-indians-diabetes.data', delimiter=',')
print(type(dataset))
# 从 txt 中导入数据，得到一个 numpy.array 
print(dataset.shape)
X = dataset[:, :8]
y = dataset[:, 8]
```

```python
seed = 7
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
print(X_train.shape)
print(y_train.shape)
print(y_train[10:20])
```

### 输入数据的形式
从上面可以看出，label用的是实值的特征，不需要进行 one-hot 处理。**而且，随便的类别 ID 都可以，不需要 0，1，...这样的顺序。**

3.训练模型

```python
model = XGBClassifier(n_estimators=50, objective='binary:logistic', silent=False)  # 全部使用默认参数简单训练一个模型
model.fit(X_train, y_train)
```

[机器学习系列(12)_XGBoost参数调优完全指南（附Python代码）](http://blog.csdn.net/han_xiaoyang/article/details/52665396)

上面输出了 xgboost 模型的一些参数:
- base_score=0.5, 
- booster='gbtree', 基本单模型，还有 gblinear, 一般 gbtree 效果更好
- colsample_bylevel=1, 用来控制树的每一级的每一次分裂，对列数的采样的占比。
- colsample_bytree=1, 控制每棵随机采样的列数的占比(每一列是一个特征)。
- gamma=0, 在节点分裂时，只有分裂后损失函数的值下降了，才会分裂这个节点。Gamma指定了节点分裂所需的最小损失函数下降值。
- **learning_rate=0.1**, 这个参数决定着每一个决定树对于最终结果（步骤2.4）的影响。GBM设定了初始的权重值之后，每一次树分类都会更新这个值，而learning_ rate控制着每次更新的幅度。通过减少每一步的权重，可以提高模型的鲁棒性。典型值为0.01-0.2。
- **max_delta_step=0,**这参数限制每棵树权重改变的最大步长。如果这个参数的值为0，那就意味着没有约束。如果它被赋予了某个正值，那么它会让这个算法更加保守。
- **max_depth=3,** 树的最大深度
- min_child_weight=1, 
- missing=None, 
- **n_estimators=100**,最大的迭代次数，也就是树的棵数。
- n_jobs=1, **注意：默认情况下是单进程**
- nthread=None,老版本，现在使用 n_job 
- objective='binary:logistic', 
- random_state=0,
- reg_alpha=0, 权重的L1正则化项。
- reg_lambda=1, 权重的L2正则化项。
- scale_pos_weight=1, 在各类别样本十分不平衡时，把这个参数设定为一个正值，可以使算法更快收敛。
- seed=None,
- silent=True, 如果 True，不会输出任何信息。典型值：0.5-1
- **subsample=1，**这个参数控制对于每棵树，随机采样的比例。

4.模型预测

```python
def my_confusion_matrix(y_true, y_pred):  
    """输出预测结果的混淆矩阵.
    http://blog.csdn.net/u010454729/article/details/50769034
    """
    from sklearn.metrics import confusion_matrix  
    labels = list(set(y_true))  
    conf_mat = confusion_matrix(y_true, y_pred, labels = labels)  
    print "confusion_matrix(left labels: y_true, up labels: y_pred):"  
    print "labels\t",  
    for i in range(len(labels)):  
        print labels[i],"\t",  
    print   
    for i in range(len(conf_mat)):  
        print i,"\t",  
        for j in range(len(conf_mat[i])):  
            print conf_mat[i][j],'\t',  
        print   
    print   
```

```python
# 对一个样本进行预测
print(X_test.shape)
sample_id = 14
y_pred0 = model.predict(X_test[sample_id].reshape([1,-1]))
print('y_pred0=', y_pred0)
y_prob0 = model.predict_proba(X_test[sample_id].reshape([1,-1]))
print('y_prob0=', y_prob0)
y_leaf0 = model.apply(X_test[sample_id].reshape([1, -1]))
print('y_leaf0=', y_leaf0)
# 树的棵数
print(y_leaf0.shape)
```

sample_id = 14

[[ 0.87595582,  0.12404419]]

现在最大深度是 3 层，那么：
- 一共会有 2^(3+1) - 1 = 15 个节点
- 叶子节点的编号范围是 [2^3-1, 2^4-1-1]，即 [7, 14]

>Q:为什么会出现编号小于 7 ？

> A:这是因为，并不是每棵树都会一直分裂，如果某个节点进行分裂后损失值下降没法超过 gamma 值的话，那么就不会继续分类。

### 预测结果
上面有3种预测方式：
- model.predict 输出预测到的类别
- model.predict_prob 输出预测为每个类别的概率值大小
- model.apply 输出该样本在每棵树中的叶子结点的 index

```python
y_leaf = model.apply(X_test)
print('y_leaf.shape=', y_leaf.shape)
# 树的棵数
min_id = np.min(y_leaf, axis=0)
max_id = np.max(y_leaf, axis=0)
min_max_id = zip(min_id, max_id)
print('min_max_id=%s' % str(min_max_id))
```

## 检测训练过程

## 调参
接下来我们用 GridSearchCV 来进行调参会更方便一些

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
```

```python
model = XGBClassifier()
learning_rate = [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3]
max_depth = [2, 3, 4]
subsample = [0.6, 0.8, 1.0]
param_grid = dict(learning_rate=learning_rate, max_depth=max_depth, subsample=subsample)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)
grid_search = GridSearchCV(model, param_grid, scoring="neg_log_loss", n_jobs=-1, cv=kfold)
grid_result = grid_search.fit(X, y)
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# 打印每个学习率对应的分数
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

寻找到最有的参数以后，还需要重新训练才能得到最优模型最后用来进行预测

```python
model = XGBClassifier(learning_rate=0.1, max_depth=3, subsample=1.0)
model.fit(X_train, y_train)

# 预测
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
